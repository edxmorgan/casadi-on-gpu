```python
import casadi as ca
from pathlib import Path
```

```python
fk_eval = ca.Function.load("fk_eval.casadi")
```

```python
base_dir = Path.cwd().parent
codegen_folder = base_dir / "src"
codegen_folder.mkdir(exist_ok=True, parents=True)

# Create a CodeGenerator object and specify the output path
cg = ca.CodeGenerator('fk_alpha',         {
        "with_header": True,
        "casadi_real": "float",
        "cpp": False,
    })
cg.add(fk_eval)
cg.generate(str(codegen_folder) + "/")
```
